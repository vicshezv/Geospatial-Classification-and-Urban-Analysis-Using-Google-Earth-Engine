```python
import rasterio
from rasterio.features import shapes
import geopandas as gpd
from shapely.geometry import shape

input_tif = r"D:\monsp\Documents\Escuela\Maestria\Eventos\CLEI 2026\02.- Recursos\Clasificacion_CDMX_recortada.tif"
output_shp = input_tif.replace(".tif", ".shp")

geometries = []
values = []

with rasterio.open(input_tif) as src:
    image = src.read(1)
    transform = src.transform
    crs = src.crs
    nodata = src.nodata

    mask = image != nodata if nodata is not None else None

    for geom, value in shapes(image, mask=mask, transform=transform):
        geometries.append(shape(geom))
        values.append(int(value))

gdf = gpd.GeoDataFrame(
    {"clase": values},
    geometry=geometries,
    crs=crs
)

gdf.to_file(output_shp, driver="ESRI Shapefile", encoding="utf-8")

print("Shapefile creado correctamente en:")
print(output_shp)
```
